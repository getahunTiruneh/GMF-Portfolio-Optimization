## Portfolio Optimization

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os, sys
# Add the 'scripts' directory to the Python path for module imports
sys.path.append(os.path.abspath(os.path.join('..', 'scripts')))
```

```python
from portfolio_optimizer import PortfolioOptimization
```

```python
tsla_data = pd.read_csv('../data/TSLA_data.csv')
bnd_data = pd.read_csv('../data/BND_data.csv')
spy_data = pd.read_csv('../data/SPY_data.csv')
```

```python
tsla_data.dropna(inplace=True)
bnd_data.dropna(inplace=True)
spy_data.dropna(inplace=True)
```

```python
# Initialize the portfolio optimization class
portfolio = PortfolioOptimization(tsla_data, bnd_data, spy_data)
```

```python
# Optimize the portfolio
optimal_weights = portfolio.optimize_portfolio()
```

```python
optimal_weights
```

```python
# Calculate the performance of the optimized portfolio
portfolio_return, portfolio_volatility, sharpe_ratio = portfolio.calculate_portfolio_performance(optimal_weights)
```

```python
# Print the optimized portfolio performance
print(f"Optimal Weights: {optimal_weights}")
print(f"Expected Portfolio Return: {portfolio_return}")
print(f"Portfolio Volatility: {portfolio_volatility}")
print(f"Sharpe Ratio: {sharpe_ratio}")
```
